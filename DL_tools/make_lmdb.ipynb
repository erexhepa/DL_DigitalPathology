```python
import lmdb
import numpy as np
import sys
import glob
import os

caffe_root = '/opt/caffe-nv/'
sys.path.insert(0, caffe_root + 'python')
import caffe
```

```python
lmdb_db_name="test_db"

in_db_label = lmdb.open(lmdb_db_name, map_size=int(1e12))
```

```python
wsize=160

for dname in xrange(1,16):
    with in_db_label.begin(write=True) as in_txn:

        print dname

        os.chdir("./"+str(dname))
        for filename in glob.glob("*.jpg"):
            print(filename)
            im_orig = caffe.io.load_image(filename)

            im_orig = caffe.io.resize_image(im_orig, [160,160])
            im_orig = im_orig[:, :, (2, 1, 0)]
            im_orig = im_orig.transpose((2, 0, 1))
            
            im_orig = img_as_ubyte(im_orig)
            
            im_dat = caffe.io.array_to_datum(im_orig)

            im_dat.label = dname-1
            key='{:0>10d}'.format(np.random.randint(0,10000000))
            key= '%s_%s'% (key, filename)
            in_txn.put(key, im_dat.SerializeToString())
    os.chdir("..")
    
```
