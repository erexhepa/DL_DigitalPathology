```python
def parallelSlideAnalysis(sequenceID):
    
    global matches
    global studyHistodistribution 
    global nmbBins 
    
    distributionHistochanels = np.zeros([1,nmbBins*2+4])
    
    #print (matches[sequenceID])
    img  = OpenSlide(matches[sequenceID])
    
    img_dimensions = (img.dimensions)        # Retourne la dimension global 
    #print(img.properties)        # Un tas de mÃ©tadata 
    img_level_count = (img.level_count)       # Retourne le nombre dâ€™Ã©tage dans la pyramide 
    img_level_dimensions = (img.level_dimensions)  # Retourne les dimensions de chaque Ã©tages
    img_level_downsamples = (img.level_downsamples) # Retourne le zoom de chaque Ã©tages

    #lowres = img.read_region(location=(10,10), level=6, size=img_level_dimensions[5])
    #lowres = img.read_region(location=(200,200), level=6, size=img_level_dimensions[7])
    lowres = img.get_thumbnail((5000,10000))

    arr = array(lowres)
    arr.shape
    
    arrDeconv = separate_stains(arr[:,:,0:3],fgx_from_rgb)
    
    # first channel statistics
    
    vecPixval = arrDeconv[:,:,0].flatten()*1.0
    vecPixvalNorm = (vecPixval - np.percentile(vecPixval,5))/np.abs(np.percentile(vecPixval,95)-np.percentile(vecPixval,5))*255
    vecPixvalNorm[np.where(vecPixvalNorm<=0)[0]]=0
    vecPixvalNorm[np.where(vecPixval>=255)[0]]=255
    foregroundPix = np.where((vecPixvalNorm>0)&(vecPixvalNorm<255))[0]
    resHistVal = np.histogram(vecPixvalNorm[foregroundPix],bins=range(0,255,2),normed=True)
    distributionHistochanels[0,:127] = resHistVal[0]
    distributionHistochanels[0,127] = np.percentile(vecPixval,5)
    distributionHistochanels[0,128] = np.percentile(vecPixval,95)

    # second channel statistics
    
    vecPixval = arrDeconv[:,:,1].flatten()*1.0
    vecPixvalNorm = (vecPixval - np.percentile(vecPixval,5))/np.abs(np.percentile(vecPixval,95)-np.percentile(vecPixval,5))*255
    vecPixvalNorm[np.where(vecPixvalNorm<=0)[0]]=0
    vecPixvalNorm[np.where(vecPixval>=255)[0]]=255
    foregroundPix = np.where((vecPixvalNorm>0)&(vecPixvalNorm<255))[0]
    resHistVal = np.histogram(vecPixvalNorm[foregroundPix],bins=range(0,255,2),normed=True)
    distributionHistochanels[0,129:256] = resHistVal[0]
    distributionHistochanels[0,256] = np.percentile(vecPixval,5)
    distributionHistochanels[0,257] = np.percentile(vecPixval,95)
    
    studyHistodistribution[sequenceID,:] = distributionHistochanels
    
    #print(resHistVal[0])
    
    return(distributionHistochanels)
```
